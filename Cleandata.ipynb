```python
df = df.rename(columns = {
    'a' : 'year',
    'b' : 'month',
    'c' : 'day',
    'd' : 'hour',
    'e' : 'minute',
    'f' : 'second',
    'g' : 'carbon_monoxide',
    'h' : 'PM_2.5',
    'i' : 'external_temp',
    'j' : 'humidity',
    'k' : 'internal_temperature',
    'l' : 'ignore'
})
df
```

```python
df = df.iloc[:,:12]
```

```python
df
```

```python
#data before cleaning
```

```python
df.plot(use_index=True,y='external_temp',color='red',figsize = (20,10))
```

```python
df.plot(use_index=True,y='ignore',color='red',figsize = (20,10))
```

```python
df.plot(use_index = True,y='humidity',color='red',figsize = (20,10))
```

```python
#clean data
df.isnull().sum()
```

```python
df = df.dropna()
```
